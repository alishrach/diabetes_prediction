```python
#dividing the data into x and y where x contains the feature columns and y contains the target column
x=df.drop(columns='Outcome',axis=1)
y=df['Outcome']
```

```python
display(x)
```

```python
display(y)
```

```python
#Standardizing the data
scaler= StandardScaler()
```

```python
standarized_data= scaler.fit_transform(x)
```

```python
display(standarized_data)
```

```python
#checking if the data is standardized
standarized_data.std()
```

```python
X=standarized_data
```

```python
#splitting the data into training and testing tests
x_train, x_test, y_train, y_test= train_test_split(X,y, test_size=0.2, stratify=y, random_state=3)
```

```python
#checking the rows and columns of the data before splitting
X.shape
```

```python
#checking the rows and columns of the training set
x_train.shape
```

```python
#checking the rows and columns of the test set
x_test.shape
```

**TRAINING THE MODEL USING SVM(SUPPORT VECTOR MACHINE**

```python
model=svm.SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

```python
#predict the values on training dataset (if a person is diabetic or not)
d_predict=model.predict(x_train)
```

```python
#accuracy of the training dataset
d_acc=accuracy_score(d_predict,y_train)
print("The accuracy for the training data is:",d_acc)
```

```python
#predict the values on testing dataset (if a person is diabetic or not)
test_predict=model.predict(x_test)
```

```python
#accuracy of the testing dataset
test_acc=accuracy_score(test_predict, y_test)
print("The accuracy score for the testing data is:",test_acc)
```
